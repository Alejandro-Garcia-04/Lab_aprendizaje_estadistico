## Ahora con train test

### Regresion

```python
X=df.drop(columns=["sales"])
y=df['sales']
x_train, x_test, y_train, y_test = train_test_split(X, y, train_size=0.6, random_state=42)
```

```python
lr = LinearRegression()
escalar=StandardScaler()
x_train=escalar.fit_transform(x_train)
lr.fit(x_train, y_train)
r2_train=r2_score(y_train, lr.predict(x_train))
print("R2 en train:", r2_train)
x_test=escalar.transform(x_test)
r2_test=r2_score(y_test, lr.predict(x_test))
print("R2 en test:", r2_test)
```

### Arbol

```python
ad=DecisionTreeRegressor()
ad.fit(x_train,y_train)
r2_train_ad=r2_score(y_train,ad.predict(x_train))
print("R2 en train del arbol:", r2_train_ad)
r2_test_ad=r2_score(y_test,ad.predict(x_test))
print("R2 en test del arbol:", r2_test_ad)
```

El arbol parece ser mejor modelo

# Polynomial Features

### Regresion lineal

```python
poly=PolynomialFeatures(degree=2)
X_poly=poly.fit_transform(X)
x_train_p, x_test_p, y_train_p, y_test_p = train_test_split(X_poly, y, train_size=0.6, random_state=42)
```

```python
lr_poly=LinearRegression()
escalar_poly=StandardScaler()
x_train_p=escalar_poly.fit_transform(x_train_p)
lr_poly.fit(x_train_p, y_train_p)
r2_train_poly=r2_score(y_train_p, lr_poly.predict(x_train_p))
print("R2 en train polinomial:", r2_train_poly)
x_test_p=escalar_poly.transform(x_test_p)
r2_test_poly=r2_score(y_test, lr_poly.predict(x_test_p))
print("R2 en test polinomial:", r2_test_poly)
```

### Arbol de decision

```python
ad_ploy=DecisionTreeRegressor()
ad_ploy.fit(x_train_p,y_train_p)
r2_train_ad_poly=r2_score(y_train_p,ad_ploy.predict(x_train_p))
print("R2 en train del arbol polinomial:", r2_train_ad_poly)
r2_test_ad_poly=r2_score(y_test_p,ad_ploy.predict(x_test_p))
print("R2 en test del arbol polinomial:", r2_test_ad_poly)
```

Con polynomial features la regresion lineal y el arbol estan al mismo nivel
