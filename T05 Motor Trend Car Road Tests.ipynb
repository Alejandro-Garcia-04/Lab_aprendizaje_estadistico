## 1.1 Realiza una regresión tomando 'mpg' como salida y eliminando la columna 'model'. Considera todos los demás factores como numéricos/ordinales.

Calcula el R2 e interpreta los signos de los betas.

Realiza un train-test-split donde se use el 40% de los datos para entrenar. Calcula el R2 de entrenamiento y de prueba.

Añade regularización L2 con un hiperparámetro lambda decidido por ti. Cambia este valor y compara con varios distintos los R2 de entrenamiento y de prueba.


```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
import seaborn as sns
from scipy import stats
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
```

```python
df=pd.read_excel('Motor Trend Car Road Tests.xlsx')
df.head()
```

```python
df.drop(columns=['model'], inplace=True)
```

### Calcula el R2 e interpreta los signos de los betas.


```python
cols=df.columns
cols=cols.drop('mpg')

x_list = []

for col in cols:
    x_list.append(df[col].values.reshape(-1, 1))


X = np.hstack(x_list)
y = df["mpg"]

scaler=StandardScaler().fit(X)
Xscaled=scaler.transform(X)

x=np.ones((len(X),1))
X_e=np.hstack((x,Xscaled))

ols=sm.OLS(y,X_e)
results=ols.fit()
results.summary()
```

### Realiza un train-test-split donde se use el 40% de los datos para entrenar. Calcula el R2 de entrenamiento y de prueba.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.4,random_state=137)
scaler=StandardScaler().fit(X_train)
Xscaled_train=scaler.transform(X_train)

x=np.ones((len(X_train),1))
X_e_train=np.hstack((x,Xscaled_train))

ols=sm.OLS(y_train,X_e_train)
results=ols.fit()
results.summary()
```

```python
Xscaled_test = scaler.transform(X_test)
x=np.ones((len(X_test),1))
X_e_test=np.hstack((x,Xscaled_test))

y_pred=results.predict(X_e_test)
r2_score(y_test,y_pred)
```

### Añade regularización L2 con un hiperparámetro lambda decidido por ti. Cambia este valor y compara con varios distintos los R2 de entrenamiento y de prueba.


```python
ridge = Ridge(alpha=10)
ridge.fit(X_e_train, y_train)
r2_train = r2_score(y_train, ridge.predict(X_e_train))
r2_test = r2_score(y_test, ridge.predict(X_e_test))
print(f'R2 Train: {r2_train}, R2 Test: {r2_test}')
```

Después de probar con varios alphas, encontré que 10 me daba el mejor valor posible, equilibrando el mejor resultado posible para ambas r2. Este fue una mejora exponencial al modelo que no tenía una regulación.

## 1.2 Repite el ejercicio anterior usando 'qsec' como salida.



#### Regresion clasica

```python
cols=df.columns
cols=cols.drop('qsec')

x_list = []

for col in cols:
    x_list.append(df[col].values.reshape(-1, 1))


X = np.hstack(x_list)
y = df["qsec"]

scaler=StandardScaler().fit(X)
Xscaled=scaler.transform(X)

x=np.ones((len(X),1))
X_e=np.hstack((x,Xscaled))

ols=sm.OLS(y,X_e)
results=ols.fit()
results.summary()
```

#### Train/Test

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.4,random_state=137)
scaler=StandardScaler().fit(X_train)
Xscaled_train=scaler.transform(X_train)

x=np.ones((len(X_train),1))
X_e_train=np.hstack((x,Xscaled_train))

ols=sm.OLS(y_train,X_e_train)
results=ols.fit()
r2_train=results.rsquared
print(f'El r cuadrado de entrenamineto es {r2_train}')
Xscaled_test = scaler.transform(X_test)
x=np.ones((len(X_test),1))
X_e_test=np.hstack((x,Xscaled_test))

y_pred=results.predict(X_e_test)
print(f'El r cuadrado de prubea es {r2_score(y_test,y_pred)}')
```

#### Regularización L2

```python
ridge = Ridge(alpha=1)
ridge.fit(X_e_train, y_train)
r2_train = r2_score(y_train, ridge.predict(X_e_train))
r2_test = r2_score(y_test, ridge.predict(X_e_test))
print(f'R2 Train: {r2_train}, R2 Test: {r2_test}')
```

Después de probar con varios alphas, encontré que 1 me daba el mejor valor posible, equilibrando el mejor resultado posible para ambas r2. Este fue una mejora exponencial al modelo que no tenía una regulación.

## 2.1 Realiza una regresión tomando 'mpg' como salida y eliminando la columna 'model'. Crea columnas dummies para los factores 'cyl', 'gear' y 'carb'.

```python
df_dummies=pd.get_dummies(df,columns=['cyl','gear','carb'])
df_dummies.head()
```

#### Train/test

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.4,random_state=137)
scaler=StandardScaler().fit(X_train)
Xscaled_train=scaler.transform(X_train)

x=np.ones((len(X_train),1))
X_e_train=np.hstack((x,Xscaled_train))

ols=sm.OLS(y_train,X_e_train)
results=ols.fit()
r2_train=results.rsquared
print(f'El r cuadrado de entrenamineto es {r2_train}')
Xscaled_test = scaler.transform(X_test)
x=np.ones((len(X_test),1))
X_e_test=np.hstack((x,Xscaled_test))

y_pred=results.predict(X_e_test)
print(f'El r cuadrado de prueba es {r2_score(y_test,y_pred)}')
```
